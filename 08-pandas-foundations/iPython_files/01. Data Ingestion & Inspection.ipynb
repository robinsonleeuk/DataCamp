# Numpy and Pandas Working Together

# Importing and Exporting Data

Sunspot data from [cilso](http://www.sidc.be/silso/infossntotdaily)


```python
filepath = "C:/Users/leero/Jupyter Notebook Directory/DataCamp/ISSN_D_tot.csv"
sunspots = pd.read_csv(filepath)
sunspots.info()

# Data is all numeric and there seem to be no missing values (there are though, as we will see)
```

The data has no headers and the columns are formatted messily. The columns shoudl be:
* Columns 0-2: Gregorian date (year, month, day)
* Column 3: Date as fraction as year
* Column 4: Daily total sunspot number
* Column 5: Definitive/provisional indicator (1 or 0)

Also, there are these problems:
* Missing values in column 4: indicated by -1
* Dates representation inconvenient

```python
# "header = None" will allow pandas to read in the csv file and recognise that there are no data headers in it

sunspots = pd.read_csv(filepath, header = None)
sunspots.head()
```

```python
# Using the info from  the cilso website on what each column represents, we can create a list of column names 

col_names = ['year', 'month', 'day', 'dec_date','sunspots', 'definite']
sunspots = pd.read_csv(filepath, header = None, names=col_names)
sunspots.head()
```

```python
# Cilso also tells us that the -1 in the sunspots column actualy means missing data as sunspots are always positive 
# if recorded. So we need to fix this. na_values is an argument you can pass to read_csv

sunspots = pd.read_csv(filepath, header = None, names=col_names, na_values='-1')
sunspots.iloc[10:20,:]
```

```python
# They are still there. Closer inspection of the csv file shows that the -1s all actually have a space in front of them, so 
# the string representing them needs to be ' -1'

sunspots = pd.read_csv(filepath, header = None, names=col_names, na_values=' -1')
sunspots.iloc[10:20,:]
```

```python
# na_values can also be done for each column, and a few different variables, by passing dictionaries and lists 
# in that argument 

sunspots = pd.read_csv(filepath, header=None,names=col_names, na_values={'sunspots':[' -1']})
sunspots.iloc[10:20,:]
```

```python
# Dates can be parsed too when you read in a csv file. Pandas reads intellgently if you give it several inputs 
# (for year, date, day etc) into the date parser

sunspots = pd.read_csv(filepath, header=None,names=col_names, na_values={'sunspots':[' -1']}, parse_dates=[[0,1,2]])
sunspots.iloc[10:20,:]
```

```python
# The 'year-month-date' column name is not ideal and also, this column would serve as an index

sunspots.index = sunspots['year_month_day']
sunspots.index.name = 'date'
sunspots.info()
```

```python
# The year_month_day and dec_date columns are still there, however, but we can get rid of those

cols = ['sunspots', 'definite']
sunspots = sunspots[cols]
sunspots.iloc[10:20,:]
```

## Exporting to csv

```python
# Importing a dataframe with the temperature in Austin, Texas

import pandas as pd
import matplotlib.pyplot as plt
temp = pd.read_csv('temp.csv', index_col=0)
temp.head()
```

## Multiple plotting

Sometimes, the data in a dataframe differs in type or scale. In these cases, plotting on the same plot will run into scaling problems. The following dataframe with weather data illustrates

```python
weather = pd.read_csv('weather.csv', index_col=0)
weather.head()
```

```python
weather.plot()
plt.show()
```

The solution is to use multiple plots by stating the 'subplots=True' argument when calling plot

```python
weather.plot(subplots=True)
plt.show()
```

Plotting a single variable or a group of variables is possible by extracting that column from the dataframe

```python
column_list1 = ['Dew Point (deg F)']
weather[column_list1].plot()
plt.show()
```

```python
column_list2 = ['Temperature (deg F)','Dew Point (deg F)']
weather[column_list2].plot()
plt.show()
```
